```python
#Write a Python script to concatenate following dictionaries to create a new one.
#Sample Dictionary :
#dic1={1:10, 2:20}
#dic2={3:30, 4:40}
#dic3={5:50,6:60}
#Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

# Define the given dictionaries
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}

# Concatenate the dictionaries
result = {}
for d in [dic1, dic2, dic3]:
    result.update(d)

# Print the concatenated dictionary
print(result)
```

```python
#7.Write a Python program to get the Fibonacci series between 0 to 50.
#Note : The Fibonacci Sequence is the series of numbers :
#0, 1, 1, 2, 3, 5, 8, 13, 21, ....
#Every next number is found by adding up the two numbers before it.
#Expected Output : 1 1 2 3 5 8 13 21 34

# initialize variables
a, b = 0, 1

# loop through the sequence
while b < 50:
    print(b, end=' ')
    # swap variables
    a, b = b, a+b
```

```python
#1. Write a Python program to add, subtract, multiple and divide two Pandas Series.
#Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

import pandas as pd

# create two Pandas Series
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 9])

# add the two series
addition = s1 + s2
print("Addition:")
print(addition)

# subtract the two series
subtraction = s1 - s2
print("Subtraction:")
print(subtraction)

# multiply the two series
multiplication = s1 * s2
print("Multiplication:")
print(multiplication)

# divide the two series
division = s1 / s2
print("Division:")
print(division)
```

# Strings

```python
#Write a Python program to concatenate all elements in a list into a string and return it.
```

```python
def concatenate_list_elements(lst):
    """
    Concatenates all elements in a list into a string and returns it.
    """
    concatenated_str = ""
    for element in lst:
        concatenated_str += str(element)
    return concatenated_str
```

```python
my_list = [1, 2, 3, "hello", "world"]
concatenated_string = concatenate_list_elements(my_list)
print(concatenated_string)
```
